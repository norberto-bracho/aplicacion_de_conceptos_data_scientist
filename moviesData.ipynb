```python
df['titulo'] = (
    df['titulo']
    .str.replace(r'\bnull\b',' ', case=False, regex=True)
    .str.replace(r'\s+',' ',regex=True)
    .str.strip()
)
```

```python
df['titulo'] = (
     df['titulo']
    .str.replace(r'\b\d{1,4}([/-]\d{1,4}){0,2}\b','',regex=True)
    .str.replace(r'\d+',' ',regex=True)
    .str.replace(r'\s+',' ',regex=True)
    .str.strip()
)
```

```python
df.sample(7)
```

```python
df['titulo'] = (
    df['titulo']
    .str.replace(r'(?i)null', ' ', regex=True)
    .str.replace(r'\b\d{1,4}([/-]\d{1,4}){0,2}\b', ' ', regex=True)
    .str.replace(r'\d+', ' ', regex=True)
    .str.replace(r'[/:.,]+', ' ', regex=True)
    .str.replace(r'(?<=[a-záéíóúñ])(?=[A-ZÁÉÍÓÚÑ])', ' ', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)
```

```python
df.sample(10)
```

```python
import re

meses = (
    'enero|febrero|marzo|abril|mayo|junio|'
    'julio|agosto|septiembre|setiembre|octubre|noviembre|diciembre'
)

df['titulo'] = (
    df['titulo']
    #  eliminar meses con o sin guión alrededor
    .str.replace(
        rf'(?i)\s*[-–—]?\s*\b({meses})\b\s*[-–—]?\s*',
        ' ',
        regex=True
    )

    # 2 normalizar espacios
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)
```

```python
df.sample(10)
```

```python
mediana = df['entradas_total'].median()
df['entradas_total']= df['entradas_total'].fillna(mediana)
```

```python
print("desvios estandar de cada varaible\n")
std_rank = df['ranking'].std()
std_semCart = df['semana_cartel'].std()
std_espSem = df['espectadores_semana'].std()
std_entTot = df['entradas_total'].std()
print("ranking:",std_rank)
print("semana cartel:",std_semCart)
print("espectadores semana:",std_espSem)
print("entradas totales:",std_entTot)
```

```python
print("varianza de cada varibale\n")
var_rank = df['ranking'].var()
var_semCart = df['semana_cartel'].var()
var_espSem = df['espectadores_semana'].var()
var_entTot = df['entradas_total'].var()
print("ranking:",var_rank)
print("semana cartel:",var_semCart)
print("espectadores semana:",var_espSem)
print("entradas totales:",var_entTot)
```

```python
print("coeficiente de variacion (CV)")
print("medias de cada variable")
media_rank = df['ranking'].mean()
media_sem = df['semana_cartel'].mean()
media_esp = df['espectadores_semana'].mean()
media_entra = df['entradas_total'].mean()

print("ranking",media_rank)
print("semana cartel",media_sem)
print("espectadores semana",media_esp)
print("entradas totales:",media_entra)

print("CV")
cv_rank = std_rank / media_rank
cv_semCart = std_semCart / media_sem
cv_espSem = std_espSem / media_esp
cv_entTot = std_entTot / media_entra

print("ranking",cv_rank)
print("semana cartel",cv_semCart)
print("espectadores semana",cv_espSem)
print("entradas totales:",cv_entTot)
```
